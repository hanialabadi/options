```python
import sys
import os
import pandas as pd
import yfinance as yf

# Print debug info for working directory and contents
print("Notebook cwd:", os.getcwd())
print("core/ dir:", os.listdir('/Users/haniabadi/Documents/Github/options/core/'))
print("chart_engine/ dir:", os.listdir('/Users/haniabadi/Documents/Github/options/core/chart_engine/'))

sys.path.append('/Users/haniabadi/Documents/Github/options')
sys.path.append('/Users/haniabadi/Documents/Github/options/core')

# Import modular engines AND feature_engine
try:
    from core.chart_engine.aggregate import aggregate_chart_tags, run_phase8_chart_engine, run_chart_verdict_on_new
    from core.chart_engine.volume_engine import compute_volume_overlays
    from core.chart_engine.pattern_engine import detect_candlestick_patterns
    from core.chart_engine.trend_engine import get_chart_trend_state
    from core.chart_engine.momentum_engine import compute_momentum_indicators
    from core.chart_engine.feature_engine import compute_all_features    # <-- add this line!
    print("‚úÖ Imported with core.chart_engine.*")
except ModuleNotFoundError as e1:
    try:
        from chart_engine.feature_engine import compute_all_features
        print("‚úÖ Imported feature_engine with chart_engine.*")
    except ModuleNotFoundError as e2:
        print("‚ùå Still can't import feature_engine.")
        print("Error1:", e1)
        print("Error2:", e2)
        raise

# --- Use modular engines AND AI feature engine for each ticker ---
tickers = enriched_with_liquidity['Underlying'].dropna().unique()

chart_results = []
for ticker in tickers:
    # Fetch historical data
    df_hist = yf.Ticker(ticker).history(period='2y', interval='1d').reset_index()
    # Core chart analytics
    df_vol = compute_volume_overlays(df_hist.copy())
    patterns, candle_data = detect_candlestick_patterns(df_hist.copy())
    trend_tag, score, trend_data = get_chart_trend_state(df_hist.copy())
    momentum = compute_momentum_indicators(df_hist.copy())
    # AI features (call with both df_hist and ticker)
    features = compute_all_features(df_hist, ticker=ticker)
    # Aggregate all results, merging the features dict in
    row = {
        'Underlying': ticker,
        'Volume_Trend': df_vol['Volume_Trend'].iloc[-1],
        'OBV': df_vol['OBV'].iloc[-1],
        'ATR': df_vol['ATR'].iloc[-1],
        'Patterns': patterns,
        'CandleData': candle_data,
        'TrendTag': trend_tag,
        'TrendScore': score,
        'TrendData': trend_data,
        'Momentum': momentum,
    }
    row.update(features)   # <-- Add all AI features as columns
    chart_results.append(row)

chart_results_df = pd.DataFrame(chart_results)
display(chart_results_df.head())

# Just before PCS
print("Before PCS:", df.columns.tolist())
print(df.shape)
display(df.head())
```
